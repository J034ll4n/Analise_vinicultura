```python
import pandas as pd
```

```python
# Lê o arquivo com separador de tabulação
df = pd.read_csv("Datasets/ProcessaViniferas.csv", sep=';', encoding='utf-8')

# Exibe as primeiras linhas para ver se a leitura deu certo
df

```

```python
# Verificando a quantidade de valores 'nd' e 'NaN' por coluna
nd_count = df.isna().sum()  # Conta os valores NaN
nd_values = df.isin(['nd']).sum()  # Conta os valores 'nd'

# Exibindo a quantidade de 'NaN' e 'nd' por coluna
print("Valores 'NaN' por coluna:")
print(nd_count)

print("\nValores 'nd' por coluna:")
print(nd_values)
```

```python
# Filtrando linhas que contêm 'NaN'
linhas_nan = df[df.isna().any(axis=1)]

# Filtrando linhas que contêm 'nd'
linhas_nd = df[df.isin(['nd']).any(axis=1)]

# Exibindo as linhas com 'NaN'
print("Linhas com valores 'NaN':")
print(linhas_nan)

# Exibindo as linhas com valores 'nd'
print("\nLinhas com valores 'nd':")
print(linhas_nd)
```

```python
# Exibe apenas as colunas com NaN para cada linha
print(df[df.isna().any(axis=1)][['id', 'control', 'cultivar', '2023']])  # Exemplo com as colunas 'id', 'control', 'cultivar' e '2023'
```

```python
# Substitui os valores 'ND' por NaN nas colunas de 2019 e 2022
df['2019'] = df['2019'].replace('ND', float('NaN'))
df['2022'] = df['2022'].replace('ND', float('NaN'))

# Exclui as linhas onde há 'NaN' em 2019 ou 2022
df = df.dropna(subset=['2019', '2022', ])

# Substitui os valores NaN na coluna 2023 por 0
df['2023'] = df['2023'].fillna(0)

# Exibe o DataFrame atualizado para verificar as mudanças
df.head()
```

```python
# Verificando a quantidade de valores 'nd' e 'NaN' por coluna
nd_count = df.isna().sum()  # Conta os valores NaN
nd_values = df.isin(['nd']).sum()  # Conta os valores 'nd'

# Exibindo a quantidade de 'NaN' e 'nd' por coluna
print("Valores 'NaN' por coluna:")
print(nd_count)

print("\nValores 'nd' por coluna:")
print(nd_values)
```

```python

# Dropar a coluna de 2019
df = df.drop(columns=['2019'])
```

```python
# Verificando a quantidade de valores 'nd' e 'NaN' por coluna
nd_count = df.isna().sum()  # Conta os valores NaN
nd_values = df.isin(['nd']).sum()  # Conta os valores 'nd'

# Exibindo a quantidade de 'NaN' e 'nd' por coluna
print("Valores 'NaN' por coluna:")
print(nd_count)

print("\nValores 'nd' por coluna:")
print(nd_values)
```

```python


# Renomear a coluna 'cultivar' para 'NOME'
df = df.rename(columns={'NOME': 'Nome'})


# Exibir as primeiras linhas para verificar se as mudanças foram feitas corretamente
df.head()
```

```python
print(df.columns)
```

```python
df = df.rename(columns={'cultivar': 'Nome'})
```

```python
# Remove colunas 'id' e 'control'
df = df.drop(columns=['id', 'control'])

# Renomeia 'produto' para 'Nome'
df = df.rename(columns={'produto': 'Nome'})

# Transforma para formato long
df_long = pd.melt(
    df,
    id_vars=['Nome'],              # Colunas fixas
    var_name='Ano',
    value_name='Quantidade (kg)'  # Nome da nova coluna de valor
)

# Converte ano para inteiro (caso esteja como string)
df_long['Ano'] = df_long['Ano'].astype(int)

# Converte quantidade para numérico
df_long['Quantidade (kg)'] = pd.to_numeric(df_long['Quantidade (kg)'], errors='coerce')

# Visualizar
print(df_long.head())
```
