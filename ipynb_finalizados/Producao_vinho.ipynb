```python
import pandas as pd
```

```python
# Lê o arquivo com separador de tabulação
df = pd.read_csv("Datasets/Producao.csv", sep=';', encoding='utf-8')

# Exibe as primeiras linhas para ver se a leitura deu certo
print(df.head())
```

```python
df.info()
```

```python
# Conta o total de valores nulos
num_nulos = df.isnull().sum().sum()
print(f"Existe(m) {num_nulos} valor(es) nulo(s).")
```

```python
df.head(10)
```

```python
# Converte de wide para long mantendo id, control, produto
df_long = df.melt(
    id_vars=['id', 'control', 'produto'],
    var_name='Ano',
    value_name='Quantidade (kg)'
)

# Converte o ano para inteiro (caso venha como string)
df_long['Ano'] = df_long['Ano'].astype(int)

# Ordena por produto e ano (opcional)
df_long = df_long.sort_values(by=['produto', 'Ano'])

# Mostra as primeiras linhas
print(df_long.head())
```

```python
df_long
```

```python
# Excluindo a Coluna Control
df_long = df_long.drop(columns=['control'])
df_long = df_long.drop(columns=['id'])
```

```python
df_long = df_long.rename(columns={'produto': 'Nome'})

df_long
```

```python
df_long.to_csv("Dataset_organizado/Producao.csv", index=False, encoding='utf-8')
```
