```python
import pandas as pd
```

```python
# Lê o arquivo com separador de tabulação
df = pd.read_csv("Datasets/Comercio.csv", sep=';', encoding='utf-8')

# Exibe as primeiras linhas para ver se a leitura deu certo
print(df.head())
```

```python
df.info()
```

```python
# Conta o total de valores nulos
num_nulos = df.isnull().sum().sum()
print(f"Existe(m) {num_nulos} valor(es) nulo(s).")
```

```python
# Exibe as linhas com valores nulos
linhas_nulas = df[df.isnull().any(axis=1)]
print(linhas_nulas)
```

```python
# Substitui o valor nulo na coluna 'control' por 'Desconhecido'
df['control'].fillna('Desconhecido', inplace=True)
```

```python
df.head(10)
```

```python
#reorganizar os dados
df_long = df.melt(id_vars=['id', 'control', 'Produto'],
                  value_vars=[str(year) for year in range(1970, 2024)],
                  var_name='Ano', value_name='Quantidade (L.)')

# Exibindo o novo DataFrame
print(df_long.head())
```

```python
df_long = df.drop(columns=['control', 'id'])
```

```python
df_long
```

```python
df_long = pd.melt(df_long,
                  id_vars=['Produto'],
                  var_name='Ano',
                  value_name='Quantidade (kg)')

df_long['Ano'] = df_long['Ano'].astype(int)
```

```python
df_long
```

```python
# Salvar sem o índice
df_long.to_csv("Datasets/Comercio_A.csv", index=False, encoding='utf-8')
```
