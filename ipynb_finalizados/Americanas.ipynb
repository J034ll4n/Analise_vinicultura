```python
import pandas as pd 
```

```python
# Lê o arquivo com separador de tabulação
df = pd.read_csv("Datasets/ProcessaAmericanas.csv", sep='\t', encoding='utf-8')

# Exibe as primeiras linhas para ver se a leitura deu certo
print(df.head())
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
print(df['2022'].unique())
print(df['2019'].unique())
```

```python
# Substituindo '*' e '+' e 'nd' por NaN
df['2022'] = df['2022'].replace(['*', '+', 'nd'], pd.NA)

# Convertendo a coluna para numérico, com valores não numéricos se tornando NaN
df['2022'] = pd.to_numeric(df['2022'], errors='coerce')
```

```python
# Substituindo '*' e '+' e 'nd' por 0
df['2019'] = df['2019'].replace(['*', '+', 'nd'], 0)

# Convertendo a coluna para numérico
df['2019'] = pd.to_numeric(df['2019'], errors='coerce')
```

```python
print(df['2022'].unique())
print(df['2019'].unique())
```

```python
print(df.dtypes)
```

```python
# Remover as colunas de 2019 e 2022
df = df.drop(columns=['2019', '2022'])

# Verificar as primeiras linhas para garantir que as colunas foram removidas
print(df.head())
```

```python
df
```

```python
# Remover as colunas 'id' e 'control'
df = df.drop(columns=['id', 'control'])

# Renomear a coluna 'cultivar' para 'Nome'
df = df.rename(columns={'cultivar': 'Nome'})

# Transformar para o formato long (melt)
df_long = df.melt(id_vars=['Nome'], var_name='Ano', value_name='Quantidade (Kg)')

# Verificar as primeiras linhas para garantir que a transformação foi feita
print(df_long.head())
```

```python
df_long
```

```python
# Salvar df_long sem o índice
df_long.to_csv("Dataset_organizado/Americanas.csv", index=False, encoding='utf-8')
```
