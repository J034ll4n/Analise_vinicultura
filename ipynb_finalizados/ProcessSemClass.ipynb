```python
import pandas as pd
```

```python
# Lê o arquivo com separador de tabulação
df = pd.read_csv("Datasets/ProcessaSemclass.csv", sep='\t', encoding='utf-8')

# Exibe as primeiras linhas para ver se a leitura deu certo
df

```

```python
# Verificando a quantidade de valores 'nd' e 'NaN' por coluna
nd_count = df.isna().sum()  # Conta os valores NaN
nd_values = df.isin(['nd']).sum()  # Conta os valores 'nd'

# Exibindo a quantidade de 'NaN' e 'nd' por coluna
print("Valores 'NaN' por coluna:")
print(nd_count)

print("\nValores 'nd' por coluna:")
print(nd_values)
```

```python

# Dropar a coluna de 2019 , 2022
df = df.drop(columns=['2019', '2022'])
```

```python
df
```

```python
# Excluir as colunas 'id' e 'control'
df = df.drop(columns=['id', 'control'])

# Renomear a coluna 'cultivar' para 'Nome'
df = df.rename(columns={'cultivar': 'Nome'})

# Organizar os dados usando melt, transformando as colunas de anos em uma coluna 'Ano' e seus valores em 'Quantidade'
df_melted = df.melt(id_vars=['Nome'], var_name='Ano', value_name='Quantidade (kg)')

# Exibir o DataFrame resultante
print(df_melted.head())
```

```python
df_melted.to_csv('Dataset_organizado/Sem_Classificacao.csv', index=False)
```
