```python

import pandas as pd

exchange = pd.read_csv('exchange.csv', skiprows=4)
inflacao = pd.read_csv('inflacao.csv', skiprows=4)
pib = pd.read_csv('Pib_por_pais.csv', skiprows=4)


print("Exchange:", exchange.columns.tolist())
print("Inflação:", inflacao.columns.tolist())
print("PIB:", pib.columns.tolist())
```

```python
exchange
```

```python

# Remover a coluna 'Unnamed: 69', que não contém dados úteis
exchange_cleaned = exchange.drop(columns=['Unnamed: 69'], errors='ignore')

# Transformar o DataFrame para o formato "long", empilhando as colunas de ano
exchange_long = exchange_cleaned.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], 
                                      var_name='Year', 
                                      value_name='Exchange Rate')

# Garantir que os valores de 'Year' e 'Exchange Rate' sejam numéricos
exchange_long['Year'] = pd.to_numeric(exchange_long['Year'], errors='coerce')
exchange_long['Exchange Rate'] = pd.to_numeric(exchange_long['Exchange Rate'], errors='coerce')

# Exibir as primeiras linhas do DataFrame transformado
exchange_long
```

```python
exchange_cleaned = exchange_long.dropna(subset=['Exchange Rate'])
```

```python
# Remover as colunas 'Indicator Name', 'Indicator Code' e 'Country Code'
exchange_cleaned = exchange_long.drop(columns=['Indicator Name', 'Indicator Code', 'Country Code'])

# Verificar as primeiras linhas para confirmar que as colunas foram removidas
print(exchange_cleaned.head())
```

```python
print(exchange_cleaned.head())
```

```python
# Filtrar apenas os dados do Brasil
brasil_data = exchange_cleaned[exchange_cleaned['Country Name'] == 'Brazil']

# Verificar as primeiras linhas para confirmar
brasil_data
```

```python
# Traduzir os nomes das colunas
exchange_cleaned = exchange_cleaned.rename(columns={
    'Country Name': 'Nome do País',
    'Year': 'Ano',
    'Exchange Rate': 'Taxa de Câmbio'
})

# Verificar as primeiras linhas para confirmar
print(exchange_cleaned.head())
```

```python
# Remover todas as linhas com valores NaN
exchange_cleaned = exchange_cleaned.dropna()


# Verificar o número de valores NaN por coluna
nan_values = exchange_cleaned.isna().sum()

# Exibir as colunas com valores NaN
print(nan_values)


```

```python
# Transformando os dados para que cada ano seja uma coluna
inflacao_long = inflacao.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
                              var_name='Year', value_name='Inflation Rate')

# Exibir as primeiras linhas após a transformação
inflacao_long
```

```python
# Remover as colunas 'Indicator Name', 'Indicator Code', 'Country Code'
inflacao_cleaned = inflacao.drop(columns=['Indicator Name', 'Indicator Code', 'Country Code'])
```

```python
# Remover as colunas 'Indicator Name', 'Indicator Code', 'Country Code'
inflacao_cleaned = inflacao.drop(columns=['Indicator Name', 'Indicator Code', 'Country Code', 'Unnamed: 69'])

# Transformar o dataframe de formato largo para longo
inflacao_long = inflacao_cleaned.melt(id_vars=['Country Name'], var_name='Year', value_name='Inflation Rate')

# Remover as linhas onde 'Inflation Rate' é NaN
inflacao_long_cleaned = inflacao_long.dropna(subset=['Inflation Rate'])

# Exibir as primeiras linhas após a transformação e limpeza
print(inflacao_long_cleaned.head())
```

```python
# Remover as colunas 'Indicator Name', 'Indicator Code', 'Country Code', 'Unnamed: 69'
pib_cleaned = pib.drop(columns=['Indicator Name', 'Indicator Code', 'Country Code', 'Unnamed: 69'])

# Transformar o dataframe de formato largo para longo
pib_long = pib_cleaned.melt(id_vars=['Country Name'], var_name='Year', value_name='GDP')

# Remover as linhas onde 'GDP' é NaN
pib_long_cleaned = pib_long.dropna(subset=['GDP'])

# Exibir as primeiras linhas após a transformação e limpeza
print(pib_long_cleaned.head())
```

```python
# Garantir que as colunas 'Year' e 'Ano' sejam do tipo int64
pib_long_cleaned['Year'] = pib_long_cleaned['Year'].astype(int)
inflacao_long_cleaned['Year'] = inflacao_long_cleaned['Year'].astype(int)
exchange_cleaned['Ano'] = exchange_cleaned['Ano'].astype(int)

# Merge dos três dataframes
df_final = pd.merge(pib_long_cleaned, inflacao_long_cleaned, on=['Country Name', 'Year'], how='inner')
df_final = pd.merge(df_final, exchange_cleaned, left_on=['Country Name', 'Year'], right_on=['Nome do País', 'Ano'], how='inner')

# Remover colunas desnecessárias após o merge
df_final = df_final.drop(columns=['Nome do País', 'Ano'])

# Exibir as primeiras linhas do dataframe final
print(df_final.head())
```

```python
import pandas as pd
crescimento = pd.read_csv('crescimento_populacao.csv', skiprows=4)
```

```python
crescimento
```

```python
# Convertendo para formato long
crescimento_long = pd.melt(crescimento, id_vars=['Country Name'], var_name='Year', value_name='Population')

# Filtrando as linhas que possuem valores numéricos na coluna 'Year'
crescimento_long['Year'] = pd.to_numeric(crescimento_long['Year'], errors='coerce')

# Remover qualquer linha onde o valor de 'Population' ou 'Year' seja NaN
crescimento_long_cleaned = crescimento_long.dropna(subset=['Population', 'Year'])

# Remover as colunas desnecessárias (como 'Country Code' e 'Unnamed: 69')
crescimento_long_cleaned = crescimento_long_cleaned.drop(columns=['Country Code', 'Unnamed: 69'], errors='ignore')

# Exibindo as primeiras linhas do DataFrame
print(crescimento_long_cleaned.head())
```

```python
# Realizando o merge entre df_final e crescimento_long_cleaned com base nas colunas 'Country Name' e 'Year'
df_final_com_crescimento = pd.merge(df_final, crescimento_long_cleaned, on=['Country Name', 'Year'], how='left')

# Exibindo as primeiras linhas do DataFrame final
print(df_final_com_crescimento.head())
```
