```python
import pandas as pd
```

```python
# Lê o arquivo com separador de tabulação
df = pd.read_csv("Datasets/ImpVinhos.csv", sep='\t', encoding='utf-8')

# Exibe as primeiras linhas para ver se a leitura deu certo
df

```

```python
df.info()
```

```python
df.isnull().sum().sum()

```

```python
# Exclui a coluna 'Id'
df = df.drop(columns=['Id'])

# Cria listas com anos
anos = [col for col in df.columns if not col.endswith('.1') and col != 'País']
anos_valor = [col for col in df.columns if col.endswith('.1')]

# Renomeia colunas com .1 para facilitar o melt
renomear = {ano: ano.split('.')[0] + '_valor' for ano in anos_valor}
df = df.rename(columns=renomear)

# Faz o melt para Quantidade
df_quantidade = df.melt(id_vars='País', value_vars=anos, var_name='Ano', value_name='Quantidade (Kg)')

# Faz o melt para Valor
df_valor = df.melt(id_vars='País', value_vars=list(renomear.values()), var_name='Ano', value_name='Valor_USD')

# Ajusta coluna 'Ano' no df_valor para bater com df_quantidade
df_valor['Ano'] = df_valor['Ano'].str.replace('_valor', '')

# Junta os dois DataFrames
df_final = pd.merge(df_quantidade, df_valor, on=['País', 'Ano'])

#  ordenar por País e Ano
df_final = df_final.sort_values(by=['País', 'Ano']).reset_index(drop=True)

# Visualizar
print(df_final.head())
```
