```python
import pandas as pd
```

```python
# Lê o arquivo com separador de tabulação
df = pd.read_csv("Datasets/ExpVinho.csv", sep='\t', encoding='utf-8')

# Exibe as primeiras linhas para ver se a leitura deu certo
print(df.head())
```

```python
df.info()
```

```python


# Conta o total de valores nulos
num_nulos = df.isnull().sum().sum()
print(f"Existe(m) {num_nulos} valor(es) nulo(s).")
df
```

```python
# Remove a coluna Id
df = df.drop(columns=["Id"])

# Lista para armazenar os dados no formato longo
dados_long = []

# Extrai os anos únicos da estrutura (sem o .1)
anos = sorted(set(int(str(col).split('.')[0]) for col in df.columns if col != 'País'))

# Loop para transformar os dados
for _, row in df.iterrows():
    pais = row["País"]
    for ano in anos:
        col_kg = str(ano)
        col_valor = f"{ano}.1"

        # Garante que as colunas existem
        if col_kg in df.columns and col_valor in df.columns:
            dados_long.append({
                "País": pais,
                "Ano": int(ano),
                "Quantidade (kg)": row[col_kg],
                "Valor_USD": row[col_valor]
            })

# Cria o novo DataFrame
df_long = pd.DataFrame(dados_long)

# Corrige tipos
df_long["Quantidade (kg)"] = pd.to_numeric(df_long["Quantidade (kg)"], errors="coerce")
df_long["Valor_USD"] = pd.to_numeric(df_long["Valor_USD"], errors="coerce")
df_long["Ano"] = df_long["Ano"].astype(int)

# Visualiza
print(df_long.head())
```
