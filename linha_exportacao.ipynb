```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os dados do arquivo CSV
df = pd.read_csv('zProjeto_Final/Datasets/exportacao.csv')
```

```python
# Carregar os dados
df_comercio = pd.read_csv("zProjeto_Final/Datasets/Comercio.csv")

# Filtrar os dados para os últimos 15 anos
ano_atual = df_comercio['Ano'].max()
anos_ultimos_15 = range(ano_atual - 14, ano_atual + 1)  # Para pegar os últimos 15 anos, incluindo o ano atual
df_ultimos_15_anos = df_comercio[df_comercio['Ano'].isin(anos_ultimos_15)]

# Agrupar os dados por 'Ano' e somar a 'Quantidade (kg)'
df_ano = df_ultimos_15_anos.groupby('Ano').agg({'Quantidade (kg)': 'sum'}).reset_index()

# Plotar gráfico de linha para 'Quantidade (kg)'
plt.figure(figsize=(10, 6))

# Gráfico de linha para quantidade exportada
plt.plot(df_ano['Ano'], df_ano['Quantidade (kg)'], label='Quantidade Exportada (kg)', marker='o', color='b')

# Adicionar título e labels
plt.title('Evolução das Exportações de Vinho ao Longo dos Últimos 15 Anos', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Quantidade (kg)', fontsize=12)

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Carregar os dados do arquivo CSV
df_producao = pd.read_csv("zProjeto_Final/Datasets/producao.csv")

# Verificar as primeiras linhas do dataframe
print(df_producao.head())

# Filtrar os dados para os últimos 15 anos
ano_atual = df_producao['Ano'].max()  # Obter o ano mais recente
anos_ultimos_15 = range(ano_atual - 14, ano_atual + 1)  # Últimos 15 anos, incluindo o ano atual

df_ultimos_15_anos = df_producao[df_producao['Ano'].isin(anos_ultimos_15)]

# Agrupar por ano e somar a quantidade de produção
df_ano = df_ultimos_15_anos.groupby('Ano')['Quantidade (kg)'].sum().reset_index()

# Calcular o aumento de produção entre os anos consecutivos
df_ano['Aumento (%)'] = df_ano['Quantidade (kg)'].pct_change() * 100  # Calcular a mudança percentual

# Mostrar os resultados
print(df_ano)

# Plotar gráfico de linha para a produção
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Gráfico de linha para quantidade produzida
plt.plot(df_ano['Ano'], df_ano['Quantidade (kg)'], label='Quantidade Produzida (kg)', marker='o', color='b')

# Adicionar título e labels
plt.title('Evolução da Produção de Vinho nos Últimos 15 Anos', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Quantidade Produzida (kg)', fontsize=12)

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()

# Plotar o aumento percentual
plt.figure(figsize=(10, 6))

# Gráfico de linha para o aumento percentual
plt.plot(df_ano['Ano'], df_ano['Aumento (%)'], label='Aumento Percentual de Produção (%)', marker='x', color='g')

# Adicionar título e labels
plt.title('Aumento Percentual de Produção de Vinho nos Últimos 15 Anos', fontsize=14)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Aumento (%)', fontsize=12)

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo CSV
df_producao = pd.read_csv("zProjeto_Final/Datasets/producao.csv")

# Verificar as primeiras linhas do dataframe
print(df_producao.head())

# Filtrar os dados para os últimos 15 anos
ano_atual = df_producao['Ano'].max()  # Obter o ano mais recente
anos_ultimos_15 = range(ano_atual - 14, ano_atual + 1)  # Últimos 15 anos, incluindo o ano atual

df_ultimos_15_anos = df_producao[df_producao['Ano'].isin(anos_ultimos_15)]

# Agrupar por tipo de uva e somar a quantidade de produção ao longo dos anos
df_uva_total = df_ultimos_15_anos.groupby('Nome')['Quantidade (kg)'].sum().reset_index()

# Calcular o aumento percentual de produção total de cada tipo de uva
df_uva_total['Aumento (%)'] = df_uva_total['Quantidade (kg)'].pct_change() * 100

# Selecionar as uvas com maiores aumentos de produção
top_uvas_crescimento = df_uva_total.sort_values(by='Aumento (%)', ascending=False)

# Exibir os resultados
print(top_uvas_crescimento)

# Plotar gráfico de barras para os tipos de uva com maior aumento percentual
plt.figure(figsize=(10, 6))

# Gráfico de barras para as uvas com maior aumento de produção
plt.bar(top_uvas_crescimento['Nome'], top_uvas_crescimento['Aumento (%)'], color='green')

# Adicionar título e labels
plt.title('Top Tipos de Uva com Maior Aumento de Produção nos Últimos 15 Anos', fontsize=14)
plt.xlabel('Tipo de Uva', fontsize=12)
plt.ylabel('Aumento (%)', fontsize=12)

# Exibir o gráfico
plt.xticks(rotation=45, ha="right")  # Para melhorar a legibilidade dos rótulos
plt.tight_layout()
plt.show()
```
